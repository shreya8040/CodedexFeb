```python
import pandas as pd
```

```python
ani=pd.read_csv('title.basics.csv')
df=pd.DataFrame(ani)
```

```python
df['startYear'] = pd.to_numeric(df['startYear'], errors='coerce')
df_anim = df[df['genres'].str.contains('Animation', na=False)][['primaryTitle', 'startYear']]
df_anim['decade'] = (df_anim['startYear'] // 10) * 10
anim_per_decade = df_anim.groupby('decade').size().reset_index(name='count')
```

```python
print(anim_per_decade.head)
```

```python
anim=anim_per_decade[:-1]
```

```python
import plotly.express as px
gr=px.histogram(anim, x='decade', y='count', title='Number of Animation Movies per Decade', color_discrete_sequence=["#affa84"])
gr.show()
```
