```python
import pandas as pd
film = pd.read_csv('title.basics.csv')
```

```python
import plotly.express as px
film.head()
```

```python

df=pd.DataFrame(film)
```

```python
df2=df.groupby('startYear').size().reset_index(name='count')
print(df2.tail(20))
```

```python
df2=df2[:-9]
```

```python
px.histogram(df2, x='startYear', y='count', title='Number of Films Released Each Year',color_discrete_sequence=["#BE7CFC"])
```
