```python
gval=fc.loc[0:23,["Year","US_at_current_prices_in_millions_Value"]]
```

```python
gval.head()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Waterfall(
    x=gval.Year,
    y=gval.US_at_current_prices_in_millions_Value,
    measure=['absolute'] + ['relative'] * (len(gval) - 1),
    connector = dict(line=dict(color="#e81a77", width=1)),
    increasing = dict(marker=dict(color="#5fe81a")),
    decreasing= dict(marker=dict(color="#cf8179")),
    
    textposition='outside'
))
```

```python
fig.show()
```

```python
fig.update_layout(
    yaxis_title='Trade Value (USD Billions)',
    xaxis_title='Year',
    title='Global Creative Economy Growth 2002–2023'
)
```
