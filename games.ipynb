```python
import pandas as pd
import plotly.express as px
```

```python
g=pd.read_csv('Video_Games.csv')
```

```python
g.head()
```

```python
df=pd.DataFrame(g)
```

```python
df2=df.groupby("Year_of_Release").size().reset_index(name="count")
```

```python
df2.head()
```

```python
px.bar(df2, x="Year_of_Release",y="count", color_discrete_sequence=["#2c99a3"], title="Number of Games Released Each Year")
```

```python
df3 = pd.read_csv(
    "games.csv",
    sep=",",
    engine="python",
    encoding="utf-8",
    on_bad_lines="skip"
)
```

```python
df3.head()
```

```python
stgnew=df3.groupby("Name").size().reset_index(name="count")
```

```python
stgnew.head()
```

```python
stgnew.tail()
```

```python
df3["Name"] = pd.to_datetime(df3["Name"], errors="coerce")
```

```python
df3.head()
```

```python
stgnew["Name"]=pd.to_datetime(stgnew["Name"], errors="coerce")
```

```python
stgnew["year"]=stgnew["Name"].dt.year
```

```python
stgfin=stgnew.groupby("year").size().reset_index(name="count")
```

```python
stgfin.head()
```

```python
stgfin=stgfin[:-1]
```

```python
px.scatter(stgfin, x="year", y="count", color_continuous_scale='Inferno', title="Number of Games Released Each Year")
```
